### Plotting func

```python
def plotmaps(input_map1,
            input_map2_name = 'input_norm',
            cmap = cmr.ember,
            title = 'NORM',
            minval = None,
            maxval = None,
             plotlog = False,
             plotdiff_not_ratio = False):
    
    path = '/xifu/home/mola/XIFU_Sims_Turbulence_NewConfig/Observation5/19p_125ks_bkg/19p_region_200/'
    
    input_map1 = input_map1.at[mask].set(np.nan)
    input_map2 = np.array(fits.getdata(path + 'input_maps/'+input_map2_name+'.fits'), dtype = 'float64')
    input_map2[mask] = np.nan
    
    if plotlog : 
        input_map1 = np.log10(input_map1)
        input_map2 = np.log10(input_map2)


    if minval == None :
        minval = min(np.nanmin(input_map1), np.nanmin(input_map2))
        print(minval)
    if maxval == None : 
        maxval = max(np.nanmax(input_map1), np.nanmax(input_map2))
        
    fig, ax = plt.subplots(1,3,figsize = (15,6), sharey = True)
          
    im = ax[1].imshow(input_map1, cmap = cmap, vmin = minval, vmax = maxval)
    plt.colorbar(im,fraction=0.046, pad=0.04)
    ax[1].set_title('INPUT ME')
    ax[1].set_xticks(())
    ax[1].set_yticks(())

    im = ax[0].imshow(input_map2, cmap = cmap, vmin = minval, vmax = maxval)
    plt.colorbar(im,fraction=0.046, pad=0.04) 
    ax[0].set_title('INPUT OLD')
    ax[0].set_ylabel(title)
    ax[0].set_xticks(())
    ax[0].set_yticks(())
    
    if plotlog or plotdiff_not_ratio : 
        diff = input_map1 - input_map2
        im = ax[2].imshow(diff, cmap = cmr.iceburn, vmin = -5, vmax = 5)
        plt.colorbar(im,fraction=0.046, pad=0.04) 
        ax[2].set_title('DIFFERENCE')
        
    else : 
        im = ax[2].imshow(input_map1/input_map2, cmap = cmr.iceburn, vmin = 0.99, vmax = 1.01)
        plt.colorbar(im,fraction=0.046, pad=0.04) 
        ax[2].set_title('RATIO')
        ax[2].set_xticks(())
        ax[2].set_yticks(())
        
    plt.subplots_adjust()
    
    #return input_map1, input_map2
```

```python
def plot_output_maps(input_map1,
            input_map2_name = 'output_norm',
            cmap = cmr.ember,
            title = 'NORM',
            minval = None,
            maxval = None,
             plotlog = False,
             plotdiff_not_ratio = False):
    
    path = '/xifu/home/mola/XIFU_Sims_Turbulence_NewConfig/Observation5/Y+42_Z+0/Y+42_Z+0_125ks/region_200/'

    input_map2 = np.array(fits.getdata(path + 'output_maps/'+input_map2_name+'.fits'), dtype = 'float64')
    
    if plotlog : 
        input_map1 = np.log10(input_map1)
        input_map2 = np.log10(input_map2)


    if minval == None :
        minval = min(input_map1.min(), input_map2.min())
    if maxval == None : 
        maxval = max(input_map1.max(), input_map2.max())
        
    fig, ax = plt.subplots(1,3,figsize = (15,6), sharey = True)
          
    im = ax[1].imshow(input_map1, cmap = cmap, vmin = minval, vmax = maxval)
    plt.colorbar(im,fraction=0.046, pad=0.04)
    ax[1].set_title('OUTPUT ME')
    ax[1].set_xticks(())
    ax[1].set_yticks(())

    im = ax[0].imshow(input_map2, cmap = cmap, vmin = minval, vmax = maxval)
    plt.colorbar(im,fraction=0.046, pad=0.04) 
    ax[0].set_title('OUTPUT OLD')
    ax[0].set_ylabel(title)
    ax[0].set_xticks(())
    ax[0].set_yticks(())
    
    if not plotdiff_not_ratio :
        im = ax[2].imshow(input_map1/input_map2, cmap = cmr.iceburn, vmin = 0.75, vmax = 1.25)
        plt.colorbar(im,fraction=0.046, pad=0.04) 
        ax[2].set_title('RATIO')
        ax[2].set_xticks(())
        ax[2].set_yticks(())
    
    else : 
        diff = input_map1 - input_map2
        im = ax[2].imshow(diff, cmap = cmr.iceburn, vmin = -50, vmax = 50)
        plt.colorbar(im,fraction=0.046, pad=0.04) 
        ax[2].set_title('DIFFERENCE')
        
    plt.subplots_adjust()
    
    #return input_map1, input_map2
```

### Bash files to run

```python
def create_bash_files(path,
                    n_nodes,
                     x_offsets,
                     y_offsets,
                     pointing_names):
    x_offsets = [x_offsets[i::n_nodes] for i in range(n_nodes)]
    y_offsets = [y_offsets[i::n_nodes] for i in range(n_nodes)]
    pointing_names = [pointing_names[i::n_nodes] for i in range(n_nodes)]

    for k in range(n_nodes):
        with open(path + 'bashfile_{}.sh'.format(k), 'w') as rsh:
            rsh.write('')
        with open(path + 'bashfile_{}.sh'.format(k), 'a') as rsh:
            rsh.write('#!/bin/sh\n\n')
            rsh.write('#SBATCH --job-name=mosaic_{} -N1  -c32\n'.format(k+1))
            rsh.write('#SBATCH --output={}/mosaic_{}.log\n'.format(path,k+1))
            rsh.write('#SBATCH --partition=xifu\n\n')
            rsh.write('dir=/xifu/home/mola/xifu_cluster_sim/clusters/obs5_reproduction/\n')
            
            x_offsets_arg = ''
            for x in x_offsets[k]:
                x_offsets_arg += ' '+ str(x)
            y_offsets_arg = ''
            for y in y_offsets[k]:
                y_offsets_arg += ' '+ str(y)
                
            pointing_names_arg = ''
            for p in pointing_names[k]:
                pointing_names_arg += ' '+ str(p)
                
            #rsh.write("X={}\n")
            #rsh.write("Y={}\n".format(y_offsets_arg))
            #rsh.write("pointing_names={}\n".format(pointing_names_arg))
            rsh.write('python3 /xifu/home/mola/xifu_cluster_sim/src/launch_pointing_nuwa.py ${dir} ')
            rsh.write('--x_offsets {} '.format(x_offsets_arg))
            rsh.write('--y_offsets {} '.format(y_offsets_arg))
            rsh.write('--pointing_names {} '.format(pointing_names_arg))
            rsh.write('--velocity_file /xifu/home/mola/xifu_cluster_sim/src/turbulent_speed_obs5.npy')
```

```python
n_nodes = 6
```

```python
create_bash_files('/xifu/home/mola/xifu_cluster_sim/clusters/obs5_reproduction/bashfiles/',
                  n_nodes,
                  x_offsets,
                  y_offsets,
                  pointing_names)
```

```python
def plot_output_maps(input_map1,
            input_map2_name = 'output_norm',
            cmap = cmr.ember,
            title = 'NORM',
            minval = None,
            maxval = None,
             plotlog = False,
             plotdiff_not_ratio = False):
    
    path = '/xifu/home/mola/XIFU_Sims_Turbulence_NewConfig/Observation5/19p_125ks_bkg/19p_region_200/'

    input_map2 = np.array(fits.getdata(path + 'output_maps/'+input_map2_name+'.fits'), dtype = 'float64')
    
    


    if minval == None :
        minval = min(np.nanmin(input_map1), np.nanmin(input_map2))
    if maxval == None : 
        maxval = max(np.nanmax(input_map1), np.nanmax(input_map2))
                
    fig, ax = plt.subplots(1,3,figsize = (15,6), sharey = True)
          
    
    if plotlog : 
        im = ax[1].imshow(input_map1, 
                          cmap = cmap, 
                          norm=LogNorm(vmin=minval, 
                              vmax=maxval))
        plt.colorbar(im,fraction=0.046, pad=0.04)
        
        im = ax[0].imshow(input_map2, cmap = cmap,
                          norm=LogNorm(vmin=minval, 
                              vmax=maxval))
        plt.colorbar(im,fraction=0.046, pad=0.04)
    else : 
        im = ax[1].imshow(input_map1, cmap = cmap, vmin = minval, vmax = maxval)
        plt.colorbar(im,fraction=0.046, pad=0.04)
        
        im = ax[0].imshow(input_map2, cmap = cmap, vmin = minval, vmax = maxval)
        plt.colorbar(im,fraction=0.046, pad=0.04)
        
    ax[1].set_title('OUTPUT ME')
    ax[1].set_xticks(())
    ax[1].set_yticks(())
    
    
    
    ax[0].set_title('OUTPUT OLD')
    ax[0].set_ylabel(title)
    ax[0].set_xticks(())
    ax[0].set_yticks(())
    
    if not plotdiff_not_ratio :
        im = ax[2].imshow(input_map1/input_map2, cmap = cmr.iceburn, vmin = 0.75, vmax = 1.25)
        plt.colorbar(im,fraction=0.046, pad=0.04) 
        ax[2].set_title('RATIO')
        ax[2].set_xticks(())
        ax[2].set_yticks(())
    
    else : 
        diff = input_map1 - input_map2
        im = ax[2].imshow(diff, cmap = cmr.iceburn, vmin = -50, vmax = 50)
        plt.colorbar(im,fraction=0.046, pad=0.04) 
        ax[2].set_title('DIFFERENCE')
        
    plt.subplots_adjust()
    
    #return input_map1, input_map2
```
