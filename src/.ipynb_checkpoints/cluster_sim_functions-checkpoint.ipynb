### Spatial Grid

```python
class SpatialGrid3D:

    def __init__(self, pixsize=10.05, shape=(58, 58), los_factor=5,
                    x_off = 0, y_off = 0, z_off = 0):
        """
        Define 3D spatial grid

        Parameters:
            pixsize (float): Projected pixel size in kiloparsecs
            shape (tuple): Size of grid on the field of view (Nx,Ny)
            los_factor (int): Factor by which the z-axis is multiplied (Nz = los_factor * Nx) 
            x_off (int) : Offset along x, in pixels
            y_off (int) : Offset along x, in pixels
            z_off (int) : Offset along x, in pixels

        """


        self.pixsize = pixsize

        x_size, y_size = shape
        z_size = int(shape[0]*los_factor)

        self.x = jnp.linspace(self.pixsize * (-(x_size - 1)/2 + x_off), self.pixsize * ((x_size - 1)/2 + x_off), x_size) + pixsize/2
        self.y = jnp.linspace(self.pixsize * (-(y_size - 1)/2 + y_off), self.pixsize * ((y_size - 1)/2 + y_off), y_size) + pixsize/2
        self.z = jnp.linspace(self.pixsize * (-(z_size - 1)/2 + z_off), self.pixsize * ((z_size - 1)/2 + z_off), z_size)
        self.X, self.Y, self.Z = jnp.meshgrid(self.x, self.y, self.z, indexing='ij', sparse=True)
        self.y_i, self.x_i = jnp.indices(shape)
        self.R = jnp.sqrt(self.X**2 + self.Y**2 + self.Z**2)
        self.shape = (x_size, y_size, z_size)
        self.volume = np.prod(self.shape)*pixsize**3
        
```

### Fourier Grid

```python
class FourierGrid3D:

    def __init__(self, spatial_grid):
        """
        Define 3D Fourier grid from spatial grid

        Parameters:
            spatial_grid (class): 3D spatial grid 
            
        """

        self.kx = fft.fftfreq(len(spatial_grid.x), d=spatial_grid.x[1] - spatial_grid.x[0])
        self.ky = fft.fftfreq(len(spatial_grid.y), d=spatial_grid.y[1] - spatial_grid.y[0])
        self.kz = fft.rfftfreq(len(spatial_grid.z), d=spatial_grid.z[1] - spatial_grid.z[0])
        KX, KY, KZ = jnp.meshgrid(self.kx, self.ky, self.kz, indexing='ij', sparse=True)
        self.K = jnp.sqrt(KX ** 2 + KY ** 2 + KZ ** 2)

        self.shape = self.K.shape
```

### Power Spectrum

```python
class KolmogorovPowerSpectrum:
    """
    Kolmogorov power spectrum
    """
    
    def __init__(self, 
                 sigma = 250., 
                 inj_scale = 300.,
                 alpha = 11/3):
        self.sigma = sigma
        self.inj_scale = inj_scale
        self.diss_scale = 1e-3
        self.alpha = alpha

    def __call__(self, k):
        r"""Kolmogorov power spectrum

        $$\mathcal{P}_{3D}(k)= \sigma^2 \frac{e^{-\left(k/k_{\text{inj}}\right)^2} e^{-\left(k_{\text{dis}}/k\right)^2} k^{-\alpha} }{\int 4\pi k^2  \, e^{-\left(k/k_{\text{inj}}\right)^2} e^{-\left(k_{\text{dis}}/k\right)^2} k^{-\alpha} \text{d} k}$$
        
        """
        


        k_inj = 1/self.inj_scale
        k_dis = 1/self.diss_scale
        
        #sigma = 10**log_sigma
        
        k_int = jnp.geomspace(k_inj/20, k_dis*20, 1000)
        norm = integrate.trapezoid(
            4*jnp.pi*k_int**3*jnp.exp(-(k_inj / k_int) ** 2) * jnp.exp(-(k_int/ k_dis) ** 2) * (k_int) ** (-self.alpha), 
            x=jnp.log(k_int)
                                  )
        res = jnp.where(k > 0, 
                        jnp.exp(-(k_inj / k) ** 2) * jnp.exp(-(k/ k_dis) ** 2) * (k) ** (-self.alpha),
                        0.
                       )

        return self.sigma**2 * res / norm
```

### Cluster Cube

```python
class ClusterCube:
    """
    Define a cube containing all the properties of the simulated cluster
    """

    def __init__(self,
                 cluster_z = 0.1,
                 xifu_pixel_size_m = 317e-6,
                 athena_focal_length = 12,
                 shape=(58, 58), 
                 los_factor=5,
                 x_off = 0, 
                 y_off = 0, 
                 z_off = 0):
        
        # Define cosmo to extract distances
        self.cosmo = LambdaCDM(H0 = 70, Om0 = 0.3, Ode0= 0.7)
        self.cluster_z = cluster_z
        
        # Pixel size projected in the sky
        self.pixsize_arcsec = (xifu_pixel_size_m/athena_focal_length * units.radian).to(units.arcsec)
        self.pixsize_kpc = (self.pixsize_arcsec * cosmo.kpc_proper_per_arcmin(cluster_z)).to(units.kpc)
        
        # Physical grid
        self.shape = shape
        self.grid = SpatialGrid3D(self.pixsize_kpc, shape, los_factor, x_off, y_off, z_off)
        
        # Fourier grid
        self.fourier_grid = FourierGrid3D(self.grid)
        
        # Get RA, Dec from physical grid
        #Â I ignore the contribution of Z, so I assume that the cluster is flat
        conversion = self.cosmo.kpc_proper_per_arcmin(cluster_z)
        coords = SkyCoord(
                            lon = cluster.grid.X * units.kpc / conversion,
                            lat = cluster.grid.Y * units.kpc / conversion,
                            frame = SkyOffsetFrame(origin=SkyCoord(ra=0, dec=0, unit='deg')),
                            ).transform_to('icrs')
        
        self.ra = coords.ra.to(units.deg).value
        self.ra = np.repeat(self.ra, repeats = shape[0], axis=0)
        
        self.dec = coords.dec.to(units.deg).value
        self.dec = np.repeat(self.dec, repeats = shape[0], axis =0)
       
    def create_density_cube(self, 
                            n0 = jnp.exp(-4.9), 
                            R_500 = 1309., 
                            r_c = jnp.exp(-2.7), 
                            gamma = 3, 
                            r_s = jnp.exp(-0.51), 
                            alpha = 0.7, 
                            beta = 0.39, 
                            eps = 2.6):
        r"""Compute the density function for a given radius.
        
        $$n_e^2(x)= n_0^2 \frac{(\frac{x}{r_c})^{-\alpha}}{(1 + (\frac{x}{r_c})^2)^{3\beta -\alpha /2}} \frac{1}{(1 + (\frac{x}{r_s})^{\gamma})^{\frac{\epsilon}{\gamma}}}$$

        Parameters:
            r (jnp.array): Radius to compute the density function in R500 units

        Returns:
            (jnp.array): Density function evaluated at the given radius in cm$^{-6}$
        """
        self.R_500 = R_500
        x = self.grid.R/R_500
        emiss = n0**2 * (x/r_c)**-alpha / (1 + (x/r_c)**2 )**(3*beta - alpha/2) / (1 + (x/r_s)**gamma)**(eps/gamma)
        #Cut emission at 5 R_500
        self.squared_density_cube = jnp.clip(emiss * jnp.heaviside(5. - x, 0), 0, 0.05**2) / 1.2

    def create_kT_cube(self,
                       M500 = 0.7,
                       T0 = 1.09,
                       rcool = jnp.exp(-4.4),
                       rt = 0.45,
                       TmT0 = 0.66,
                       acool = 1.33,
                       c2 = 0.3,
                       ):
        r"""Compute the temperature function for a given radius.

        $$\dfrac{T(x)}{T_{500}} = T_0 \dfrac{\frac{T_\mathrm{min}}{T_0} + (\frac{x}{r_\mathrm{cool}})^{a_\mathrm{cool}}}{1 + (\frac{x}{r_\mathrm{cool}})^{a_\mathrm{cool}}} \frac{1}{(1 + (\frac{x}{r_t})^2)^{\frac{c}{2}}}$$

        Parameters:
            r (jnp.array): Radius at which to compute the temperature, in kpc
            z (float): Redshift of cluster
        Returns:
            (jnp.array): Temperature function evaluated at the given radius in keV
        """

        # Scale T_500 to redshift
        Ez   = self.cosmo.efunc(self.cluster_z)
        T_500 = 8.85 * (M500*self.cosmo.h)**(2./3.) * Ez**(2./3.)

        # Switch to units of R_500
        x = self.grid.R/self.R_500

        
        term1 = (TmT0 + (x / rcool) ** acool)
        term2 = (1 + (x / rcool) ** acool) * (1 + (x / rt) ** 2) ** c2

        self.kT_cube = T_500 * T0 * term1 / term2 * jnp.heaviside(5. - x, 0)
        
    def create_Z_cube(self):
        """
        Compute the abundance function for a given radius, following Mernier et al 2017.

        Parameters:
            r (jnp.array): Radius to compute the temperature in kpc

        Returns:
            (jnp.array): Abundance function evaluated at the given radius
        """

        x = self.grid.R/self.R_500
        self.Z_cube = 0.21*(x+0.021)**(-0.48) - 6.54*jnp.exp( - (x+0.0816)**2 / (0.0027) )* jnp.heaviside(5. - x, 0)
                
        
    def create_norm_cube(self):
        """
        Compute the norm for the xspec apec model

        $$\text{norm} = \frac{1}{4 \pi (D_A(1+z))^2 \int n_e n_H dV}$$

        """
        
        ang_diam_distance_cm = self.ang_diam_distance_kpc.to(units.cm)
        
        xspec_normalization = 1e-14/(4*np.pi*(ang_diam_distance_cm*(1+self.cluster_z))**2)
        dV = (self.grid.pixsize.to(units.cm).value)**3
        ne_nH_dV = self.squared_density_cube * dV
        
        self.norm = xspec_normalization.value * ne_nH_dV
        
    def create_velocity_cube(self, 
                             rng_key,
                             sigma = 250., 
                             inj_scale = 300.,
                             alpha = 11/3):
        """
        Returns:
            field_spatial (jnp.array): Random realization of GRF

        """
        self.power_spectrum = KolmogorovPowerSpectrum(sigma = 250., 
                 inj_scale = 300.,
                 alpha = 11/3)
        
        self.K = self.fourier_grid.K.astype(np.float64)
                
        #Dont mind the rfft, it is here to gain memory
        field_spatial = random.normal(rng_key, shape=self.grid.shape)
        field_fourier = fft.rfftn(field_spatial)*jnp.sqrt(self.power_spectrum(self.K)/self.grid.pixsize**3)
        self.v_cube = fft.irfftn(field_fourier, s=self.grid.shape)
        
        
    def save_cube(self, path):
        """
        Saves the cubes of computed value of the cluster
        """
        np.savez_compressed(path,
                            {'X_coord_cube':self.grid.X,
                             'Y_coord_cube':self.grid.Y,
                             'Z_coord_cube':self.grid.Z,
                             'ra_cube':self.ra,
                             'dec_cube':self.dec,
                             'ne_nH_cube':self.squared_density_cube,
                             'norm_cube':self.norm,
                             'kT_cube':self.kT_cube,
                             'Z_cube':self.Z_cube,
                             'v_cube':self.v_cube})
        
    def create_input_maps(self,
                          X_pixels, 
                          Y_pixels, 
                          bin_num_pix,
                          nb_bins, 
                          PSF_kernel,
                          save_maps = False,
                          path_save = './'):
        """
        Computes the input maps from the cluster cubes
        
        Parameters:
            X_pixels (jnp.array): Array of x coordinate of each pixel on the xifusim images
            Y_pixels (jnp.array): Array of y coordinate of each pixel on the xifusim images 
            bin_num_pix (jnp.array): Array of the bin number of each pixel
            nb_bins (int): Nb of bins in the binning
            PSF_kernel (jnp.array): PSF discretized on the pixel grid
            save_maps (bool): Whether to save the input maps in a npz file
            path_save (str): Path where to save the input maps
        """
        
        # Summing in each pixel
        summed_norm = jnp.sum(self.norm, axis = -1)
        summed_kT = jnp.sum(self.kT_cube*self.norm, axis = -1)
        summed_Z = jnp.sum(self.Z_cube*self.norm, axis = -1)
        summed_v = jnp.sum(self.v_cube*self.norm, axis = -1)
        summed_std = jnp.sum(self.v_cube**2*self.norm, axis = -1)

        # Convolution by PSF
        PSF_kernel*= 1/np.sum(PSF_kernel)
        summed_norm_conv = signal.convolve(summed_norm, PSF_kernel, mode = 'same')
        summed_kT_conv = signal.convolve(summed_kT, PSF_kernel, mode = 'same')
        summed_Z_conv = signal.convolve(summed_Z, PSF_kernel, mode = 'same')
        summed_v_conv = signal.convolve(summed_v, PSF_kernel, mode = 'same')
        summed_std_conv = signal.convolve(summed_std, PSF_kernel, mode = 'same')
        
        # Indices of which bin each pixel goes in
        bins_unique, inverse_indices = jnp.unique(bin_num_pix, 
                                                  return_inverse=True, 
                                                  size = jnp.max(bin_num_pix).astype(int))
        
        # Initialize vectors for summation in bins
        summed_norm_vec = jnp.zeros(nb_bins)
        summed_kT_vec = jnp.zeros(nb_bins)
        summed_Z_vec = jnp.zeros(nb_bins)
        summed_v_vec = jnp.zeros(nb_bins)
        summed_std_vec = jnp.zeros(nb_bins)
        
        # We add to each bin the weighted sum of all pixels belonging to it
        summed_norm_vec = summed_norm_vec.at[inverse_indices].add(
                                                                summed_norm_conv[X_pixels,Y_pixels]
                                                                )
        summed_kT_vec = summed_kT_vec.at[inverse_indices].add(
                                                                summed_kT_conv[X_pixels,Y_pixels]
                                                                )
        summed_Z_vec = summed_Z_vec.at[inverse_indices].add(
                                                                summed_Z_conv[X_pixels,Y_pixels]
                                                                )
        summed_v_vec = summed_v_vec.at[inverse_indices].add(
                                                                summed_v_conv[X_pixels,Y_pixels]
                                                                )
        summed_std_vec = summed_std_vec.at[inverse_indices].add(
                                                                summed_std_conv[X_pixels,Y_pixels]
                                                                )
        
        # Divide by weighing (i.e. summed emission)
        weight_for_norm_vec = jnp.zeros(nb_bins)
        weight_for_norm_vec = weight_for_norm_vec.at[inverse_indices].add(
                                                                            jnp.ones(self.shape)[X_pixels,Y_pixels]
                                                                        )
        summed_norm_vec_weighted = summed_norm_vec[inverse_indices] / weight_for_norm_vec[inverse_indices]
        summed_kT_vec_weighted = summed_kT_vec[inverse_indices]/summed_norm_vec[inverse_indices]
        summed_Z_vec_weighted = summed_Z_vec[inverse_indices]/summed_norm_vec[inverse_indices]
        summed_v_vec_weighted = summed_v_vec[inverse_indices]/summed_norm_vec[inverse_indices]
        summed_std_vec_weighted = jnp.sqrt(summed_std_vec[inverse_indices]/summed_norm_vec[inverse_indices] - summed_v_vec_weighted**2)


        
        # Create maps
        input_norm = jnp.zeros(self.shape)
        input_norm = input_norm.at[X_pixels, Y_pixels].set(summed_norm_vec_weighted)
        
        input_kT = jnp.zeros(self.shape)
        input_kT = input_kT.at[X_pixels, Y_pixels].set(summed_kT_vec_weighted)
        
        input_Z = jnp.zeros(self.shape)
        input_Z = input_Z.at[X_pixels, Y_pixels].set(summed_Z_vec_weighted)
    
        input_v = jnp.zeros(self.shape)
        input_v = input_v.at[X_pixels, Y_pixels].set(summed_v_vec_weighted)
        
        input_std = jnp.zeros(self.shape)
        input_std = input_std.at[X_pixels, Y_pixels].set(summed_std_vec_weighted)
        
        if save_maps:
            np.savez_compressed(path_save + 'inputs_maps.npz', 
                               {'input_norm':input_norm,
                                'input_kT':input_kT,
                                'input_Z':input_Z,
                                'input_v':input_v,
                                'input_std':input_std})
    
        return input_norm, input_kT, input_Z, input_v, input_std
```

```python
from joblib import Parallel, delayed
import contextlib
import joblib
from tqdm import tqdm

@contextlib.contextmanager
def tqdm_joblib(tqdm_object):
    """Context manager to patch joblib to report into tqdm progress bar given as argument"""
    class TqdmBatchCompletionCallback(joblib.parallel.BatchCompletionCallBack):
        def __call__(self, *args, **kwargs):
            tqdm_object.update(n=self.batch_size)
            return super().__call__(*args, **kwargs)

    old_batch_callback = joblib.parallel.BatchCompletionCallBack
    joblib.parallel.BatchCompletionCallBack = TqdmBatchCompletionCallback
    try:
        yield tqdm_object
    finally:
        joblib.parallel.BatchCompletionCallBack = old_batch_callback
        tqdm_object.close()
```

```python
class PhotonList:
    """
    Create the photon lists associated 
    with the observation of a cluster
    """
    def __init__(self, 
                 cluster, 
                 E_bounds, 
                 Z_scale, 
                 T_scale, 
                 interp_spec_table,
                 exposure = 125e3,
                 exposure_overhead = 10,
                 eff_area = 2e4):
        
        self.cluster = cluster
        self.E_bounds = E_bounds
        self.E = (E_bounds[1:]+E_bounds[:-1])/2
        self.T_scale = T_scale
        self.Z_scale = Z_scale
        self.interp_spec_table = interp_spec_table
        self.exposure = exposure
        self.exposure_overhead = exposure_overhead
        self.nb_cells = np.prod(cluster.grid.shape)
        self.cluster_z = cluster.cluster_z
        self.eff_area = eff_area
              
        
    def write_photon_list_multiproc(self,numproc):
        '''
        
        '''   
        
        n_test = 50_000 #len(self.cluster.ra.flatten())
        
        args_list = zip(
                        np.array(self.cluster.ra.flatten()[:n_test]),
                        np.array(self.cluster.dec.flatten()[:n_test]),
                        np.array(self.cluster.norm.flatten()[:n_test]),
                        np.array(self.cluster.kT_cube.flatten()[:n_test]),
                        np.array(self.cluster.Z_cube.flatten()[:n_test]),
                        np.array(self.cluster.z_cube.flatten()[:n_test])
                        )
        
        if numproc > 1:
            batch_size = max(1, n_test//(4*numproc))
        else :
            batch_size = 'auto'
        
        results = Parallel(n_jobs=numproc, 
                           backend="loky", 
                           batch_size = batch_size
                          )(
                                delayed(self.add_photons_interpTZ_new)(ra, dec, norm, kT, Z, z) for ra, dec, norm, kT, Z, z in tqdm_notebook(args_list, 
                                                                                                                                           total = n_test, 
                                                                                                                                           desc = 'Creating photon list'
                                                                                                                                              )
                                )
                                                                
        
        results = [res_item for res_item in results if res_item is not None] 

        return np.hstack(results)
    
        
    def add_photons_interpTZ_new(self,
                                 ra,
                                 dec,
                                 norm,
                                 temperature,
                                 abundance,
                                 z_deltaz):
        '''
        -flux_table needed to choose the right spectra and have the distribution fonctions of the photons in function of energy
        -Z_scale, T_scale needed to interpolate the spectra in flux_table_z.npy
        -ebounds : step in energy of the spectra
        -ra_list, dec_list, energy_list : output of the function
        -ra : of the cellule ( in input_table )
        -dec : of the cellule ( in input_table )
        -temperature : of the cellule ( in input_table )
        -abundance : of the cellule ( in input_table )
        -z_deltaz : REDSHIFT WITH WHICH WE HAVE BUILT THE INPUT TABLE
        -z : redshift of the cellule
        - norm : of the cellule
        - eff_area : area of the telescope
        - exposure : exposure time
        Adds photons to the list - faster
        For memory optimisation mainly
        For a model where temperature and abundance are not fixed
        /!\ z is the redshift of the spectrum that we DO NOT shift ( the cell redshift)
        z_deltaz is the redshift of the spectrum that WE DO SHIFT!!! ( the cluster redshift, with which we built the spectrum table )
        Ex: if i want to shift my cube at 0.1 to 0.11 then z_deltaz=0.1 and z=0.11
        '''
        emin=self.E_bounds[:-1]
        emax=self.E_bounds[1:]
        #Compute number of photons for this cell
        indT=np.digitize(temperature,self.T_scale)-1
        indZ=np.digitize(abundance,self.Z_scale)-1

        #Interpolate cdf from table
        #Table : 3D array, [Z,T,flux]
        dT=temperature-T_scale[indT]
        deltaT=T_scale[indT+1]-T_scale[indT]
        dZ=abundance-Z_scale[indZ]
        deltaZ=Z_scale[indZ+1]-Z_scale[indZ]

        #print(T_scale[indT],T_scale[indT+1],Z_scale[indZ], Z_scale[indZ+1])
        
        delta_fluxT=self.interp_spec_table[indZ][indT+1]-self.interp_spec_table[indZ][indT]
        delta_fluxZ=self.interp_spec_table[indZ+1][indT]-self.interp_spec_table[indZ][indT]
        delta_fluxTZ=self.interp_spec_table[indZ][indT]+self.interp_spec_table[indZ+1][indT+1]-self.interp_spec_table[indZ][indT+1]-self.interp_spec_table[indZ+1][indT]    

        #spectre interpole dans la table de spectre
        flux_inter=(delta_fluxT*dT/deltaT + delta_fluxZ*dZ/deltaZ + delta_fluxTZ*(dT/deltaT)*(dZ/deltaZ) + self.interp_spec_table[indZ][indT])*norm

        #to shift the spectrum with the redshift if needed
        if self.cluster.cluster_z!=z_deltaz:
            Edec = self.E+self.E*(z_deltaz-self.cluster_z)/(1+self.cluster_z)
            flux_inter = np.interp(self.E, Edec, flux_inter)

        #flux du spectre, a calculer pour deccterminer le nombre de photon
        #M.V. xspec
        flux_spectre=np.trapz(flux_inter)
        #return flux_inter

        #flux en ergs.cm-2 for the simput file
        flux_erg=np.trapz(flux_inter*Edec)*units.keV.to(units.erg)

        #calcul du nombre de photons
        nphotons = flux_spectre * self.eff_area * self.exposure * self.exposure_overhead
        nphotons = np.random.poisson(nphotons)
        
        #Loop over cells and dump photons in the list
        
        
        if nphotons:
            #Find the interpoaltion index

            flux_cdf=np.cumsum(flux_inter/np.sum(flux_inter))

            #Affect an energy to each photon
            random_numbers = np.random.uniform(size=2*nphotons)
            search_indexes = np.searchsorted(flux_cdf,random_numbers[:nphotons]*flux_cdf[-1])

            energy_list = emin[search_indexes] + random_numbers[nphotons:]*(emax[search_indexes]-emin[search_indexes])
            flux_erg_list = flux_erg*np.ones(nphotons)/nphotons
            ra_list = ra*np.ones(nphotons)
            dec_list = dec*np.ones(nphotons)
            return np.vstack((flux_erg_list, energy_list, ra_list, dec_list))

        #else : 
        #    return None

```

```python
def write_simput_file(ra, dec, energy, emin, emax, flux, simputfile, clobber=False):
    r"""
    
    Write events to a SIMPUT file that may be read by the SIMX instrument
    simulator. See the SIMPUT definition for reference: 
    http://hea-www.harvard.edu/heasarc/formats/simput-1.1.0.pdf
    
    This script requires NumPy (http://www.numpy.org) and AstroPy 
    (http://www.astropy.org) to be installed in your Python distribution. 
    
    Parameters
    ----------
    ra : list or NumPy array of floats
        The RA positions of the photons in degrees.
    dec : list or NumPy array of floats
        The Dec positions of the photons in degrees.
    energy : NumPy array of floats
        The energies of the photons in keV. 
    emin : float
        The minimum energy of the photons in keV in the reference energy band. 
    emax : float
        The maximum energy of the photons in keV in the reference energy band.
    flux : float
        Total flux of photons in erg/s/cm**2 in the reference energy band.
    simputfile : string
        Name of the simput file to write
    clobber : boolean, optional
        Set to True to overwrite previous files.
    """
    #Construct photon list
    ra = np.array(ra)
    dec = np.array(dec)
    energy = np.array(energy).astype('float64')

    col1 = fits.Column(name='ENERGY', format='E', array=energy)
    col2 = fits.Column(name='RA', format='D', array=ra)
    col3 = fits.Column(name='DEC', format='D', array=dec)

    coldefs = fits.ColDefs([col1, col2, col3])
    
    #tbhdu=fits.new_table(coldefs)
    tbhdu = fits.BinTableHDU.from_columns(coldefs)
    tbhdu.name = "PHLIST"

    tbhdu.header["HDUCLASS"] = "HEASARC/SIMPUT"
    tbhdu.header["HDUCLAS1"] = "PHOTONS"
    tbhdu.header["HDUVERS"] = "1.1.0"
    tbhdu.header["EXTVER"] = 1
    tbhdu.header["REFRA"] = 0.0
    tbhdu.header["REFDEC"] = 0.0
    tbhdu.header["TUNIT1"] = "keV"
    tbhdu.header["TUNIT2"] = "deg"
    tbhdu.header["TUNIT3"] = "deg"
    
    #Construct source catalog    
    col1 = fits.Column(name='SRC_ID', format='J', array=np.array([1]).astype("int32"))
    col2 = fits.Column(name='RA', format='D', array=np.array([0.0]))
    col3 = fits.Column(name='DEC', format='D', array=np.array([0.0]))
    col4 = fits.Column(name='E_MIN', format='D', array=np.array([float(emin)]))
    col5 = fits.Column(name='E_MAX', format='D', array=np.array([float(emax)]))
    col6 = fits.Column(name='FLUX', format='D', array=np.array([flux]))
    col7 = fits.Column(name='SPECTRUM', format='80A', 
                         array=np.array(["[PHLIST,1]"]))
    col8 = fits.Column(name='IMAGE', format='80A', 
                         array=np.array(["[PHLIST,1]"]))
                        
    coldefs = fits.ColDefs([col1, col2, col3, col4, col5, col6, col7, col8])
        
    #wrhdu=fits.new_table(coldefs)
    wrhdu = fits.BinTableHDU.from_columns(coldefs)
    wrhdu.name = "SRC_CAT"
                                
    wrhdu.header["HDUCLASS"] = "HEASARC"
    wrhdu.header["HDUCLAS1"] = "SIMPUT"
    wrhdu.header["HDUCLAS2"] = "SRC_CAT"
    wrhdu.header["HDUVERS"] = "1.1.0"
    wrhdu.header["RADECSYS"] = "FK5"
    wrhdu.header["EQUINOX"] = 2000.0
    wrhdu.header["TUNIT2"] = "deg"
    wrhdu.header["TUNIT3"] = "deg"
    wrhdu.header["TUNIT4"] = "keV"
    wrhdu.header["TUNIT5"] = "keV"
    wrhdu.header["TUNIT6"] = "erg/s/cm**2"

    primary_hdu = fits.PrimaryHDU()
    thdulist = fits.HDUList([primary_hdu, wrhdu,tbhdu])
    thdulist.writeto(simputfile,overwrite=clobber)
```

```python
def merge_evt_files_recursive(output_file,pattern,clobber=False,file_list=None,depth=0):
    '''Merges different event files matching a given pattern recursively 
    
    Arguments:
        - output_file: name of the final event list to write
        - pattern: pattern to find the event files to merge
        - clobber: standard FITS clobber option
        - file_list: list of files to merge (for recursive calls)
    '''
    if os.path.exists(output_file) and not clobber:
        raise RuntimeError("File %s already exists!" % output_file)
    if file_list==None:
        file_list = glob(pattern)
        
    nb_files = len(file_list)
    if nb_files==1:
        os.system("cp -r " + file_list[0] + " " + output_file)
    else:
        if nb_files>3:
            outfile1 = os.path.dirname(output_file)+'/tmp_evt_%d_0.fits' % depth
            merge_evt_files_recursive(outfile1,'',clobber=True,file_list=file_list[:int(nb_files/2)],depth=depth+1)
            outfile2 = os.path.dirname(output_file)+'/tmp_evt_%d_1.fits' % depth
            merge_evt_files_recursive(outfile2,'',clobber=True,file_list=file_list[int(nb_files/2):],depth=depth+1)
            print('Merging',outfile1,outfile2)
            os.system("fmerge \"%s %s\" %s \" \" clobber=yes" % (outfile1,outfile2,output_file))
        elif nb_files>0:
            print("Merging",file_list[0],file_list[1])
            os.system("fmerge \"%s %s\" %s \" \" clobber=yes" % (file_list[0],file_list[1],output_file))
            for evt_file in file_list[2:]:
                print('Merging',output_file,evt_file)
                os.system("fmerge \"%s %s\" %s \" \" clobber=yes" % (output_file,evt_file,output_file))
        if depth==0:
            for tmp_file in glob(os.path.dirname(output_file)+"/tmp_evt_*_*.fits"):
                os.remove(tmp_file)
```
