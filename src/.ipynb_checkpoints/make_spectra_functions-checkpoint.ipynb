```python
def screen_events(hdu,correct_grades,pixid=None,binning_dict=None,src_id=None):
    '''
    Screens the events by getting rid of misreconstructed events
    Arguments:
        - hdu: input fits list
        - correct_grades: list containing the grades to accept
        - exclude_pileup: option to purposely exclude pileup events (this should only be done knowingly)
        - binning_dict: dict object containing the region to PIXID association
        - src_id: if given, only use events corresponding to the given src_id
    '''
    
    #Remove misreconstructed events
    data = hdu['EVENTS'].data
    want_band = is_in_list(data['GRADING'], correct_grades)
    print(hdu)
    if pixid!=None:
        if binning_dict!=None:
            want_band*=is_in_list(data['PIXID'], binning_dict[pixid][0])
        else:
            want_band*=(data['PIXID']==pixid)
    print(np.sum(want_band))
    if src_id!=None:
        want_band*=(data['SRC_ID']==src_id)
            
            
    new_hdu = fits.BinTableHDU(data=data[want_band],header=hdu[1].header)
    
    return fits.HDUList([hdu[0],new_hdu,hdu[2]])

def is_in_list(array_to_test,comparison_list):
    '''
    Returns a boolean array telling for each element of array_to_test whether it is in comparison_list
    Is typically efficient if comparison_list is small and array_to_test is big. If the other way around, use is_in_big_list 
    '''
    wb = np.zeros(len(array_to_test))
    for el in comparison_list:
        wb+=(array_to_test==el)
    return wb>=1 
```

I need to have a function of vignetting that is continuous in energy, and the issue is that the vignetting file is providing values that are integrated in energy. I approximate the vignetting with a piecewise polynomial fit, such that in each energy bin, the integral is equal to that of the corresponding value within the vignetting file. For the theta dependence, I'll just assume a linear interpolation.

```python
from scipy.linalg import solve

def quadratic_from_bin_averages(E_edges, V_row, boundary='natural', reg_eps=1e-12):
    """
    Build piecewise quadratic per bin with exact bin averages and C1 continuity.
    boundary: 'zero_slope', 'natural', or 'extrapolate'.
      - 'zero_slope': enforce derivative = 0 at both ends
      - 'natural'   : enforce second derivative = 0 at both ends (a_0 = a_last = 0)
      - 'extrapolate': set right-end derivative from trend of last bins (left derivative = 0)
    """
    E_edges = np.asarray(E_edges, dtype=float)
    V_row = np.asarray(V_row, dtype=float)
    n_bins = len(V_row)
    n_unknowns = 3 * n_bins

    # prepare system
    A = np.zeros((n_unknowns, n_unknowns), dtype=float)
    rhs = np.zeros(n_unknowns, dtype=float)
    row_idx = 0

    def avg_basis(Î):
        return np.array([Î**2 / 3.0, Î / 2.0, 1.0], dtype=float)

    # 1) bin-average constraints
    for i in range(n_bins):
        Ei = E_edges[i]; Ej = E_edges[i+1]; Î = Ej - Ei
        A[row_idx, 3*i:3*i+3] = avg_basis(Î)
        rhs[row_idx] = V_row[i]
        row_idx += 1

    # 2) continuity constraints (value + slope) between bins
    for i in range(n_bins - 1):
        Ei = E_edges[i]; Ej = E_edges[i+1]; Î = Ej - Ei
        # value continuity: a_i*Î^2 + b_i*Î + c_i - c_{i+1} = 0
        A[row_idx, 3*i:3*i+3] = [Î**2, Î, 1.0]
        A[row_idx, 3*(i+1) + 2] = -1.0
        rhs[row_idx] = 0.0
        row_idx += 1
        # slope continuity: 2*a_i*Î + b_i - b_{i+1} = 0
        A[row_idx, 3*i:3*i+3] = [2.0*Î, 1.0, 0.0]
        A[row_idx, 3*(i+1) + 1] = -1.0
        rhs[row_idx] = 0.0
        row_idx += 1

    # 3) boundary conditions
    if boundary == 'zero_slope':
        # left derivative b_0 = 0
        A[row_idx, 1] = 1.0
        rhs[row_idx] = 0.0
        row_idx += 1
        # right derivative at end: 2*a_last*Î_last + b_last = 0
        Î_last = E_edges[-1] - E_edges[-2]
        A[row_idx, -3:] = [2.0*Î_last, 1.0, 0.0]  # FIXED
        rhs[row_idx] = 0.0
        row_idx += 1

    elif boundary == 'natural':
        # natural: second derivative = 0 at left and right -> a_0 = 0, a_last = 0
        A[row_idx, 0] = 1.0
        rhs[row_idx] = 0.0
        row_idx += 1
        A[row_idx, -3] = 1.0
        rhs[row_idx] = 0.0
        row_idx += 1

    elif boundary == 'extrapolate':
        # left derivative = 0
        A[row_idx, 1] = 1.0
        rhs[row_idx] = 0.0
        row_idx += 1
        # estimate slope at right
        centers = 0.5*(E_edges[:-1] + E_edges[1:])
        if len(centers) >= 2:
            slope_est = (V_row[-1] - V_row[-2]) / (centers[-1] - centers[-2])
        else:
            slope_est = 0.0
        Î_last = E_edges[-1] - E_edges[-2]
        A[row_idx, -3:] = [2.0*Î_last, 1.0, 0.0]  # FIXED
        rhs[row_idx] = slope_est
        row_idx += 1

    else:
        raise ValueError("boundary must be 'zero_slope', 'natural', or 'extrapolate'")

    assert row_idx == n_unknowns, f"constructed {row_idx} rows but need {n_unknowns}"

    # Solve robustly
    try:
        sol = np.linalg.solve(A, rhs)
    except np.linalg.LinAlgError:
        ATA = A.T @ A
        ATb = A.T @ rhs
        ATA_reg = ATA + reg_eps * np.eye(ATA.shape[0])
        try:
            sol = np.linalg.solve(ATA_reg, ATb)
        except np.linalg.LinAlgError:
            sol, *_ = np.linalg.lstsq(A, rhs, rcond=None)

    coeffs = sol.reshape(n_bins, 3)
    return coeffs

def eval_quadratic_piecewise(E_edges, coeffs, E_eval):
    E_edges = np.asarray(E_edges, dtype=float)
    E_eval = np.asarray(E_eval, dtype=float)
    vals = np.zeros_like(E_eval, dtype=float)
    for i in range(len(coeffs)):
        a,b,c = coeffs[i]
        left = E_edges[i]; right = E_edges[i+1]
        mask = (E_eval >= left) & (E_eval <= right)
        if not np.any(mask):
            continue
        x = E_eval[mask] - left
        vals[mask] = a*x**2 + b*x + c
    vals[(E_eval < E_edges[0]) | (E_eval > E_edges[-1])] = np.nan
    return vals
```

```python
import subprocess

from joblib import Parallel, delayed
import joblib

from tqdm.notebook import tqdm as tqdm_notebook
```

```python
class MakeSpectra:
    def __init__(self,
                binning):
        self.binning = binning
        
    def load_vignetting(self,
                       vign_file_path = '/xifu/usr/share/sixte/instruments/athena-xifu_2024_11/baseline/instdata/athena_vig_13rows_20240326.fits',
                       arf_file_path = '/xifu/home/mola/XIFU_Sims_Turbulence_NewConfig/ARF.arf'):
        
        # Load vignetting file and assign arrays
        hdu_vign=fits.open(vign_file_path)
        vign=hdu_vign[1].data
        self.vign_E_centers = (vign['ENERG_LO'] + vign['ENERG_HI'])[0]/2
        self.vign_E_widths = (vign['ENERG_HI'] - vign['ENERG_LO'])[0]
        self.vign_E_bins = np.hstack((vign['ENERG_LO'][0], vign['ENERG_HI'][0][-1]))
        self.vignet_integre = np.array(vign['VIGNET'][0,0])
        self.theta = np.array(vign['THETA'][0])*60. #Arcminutes
        
        # Load arf file
        self.hdu_arf=fits.open(arf_file_path)
        arf = self.hdu_arf[1].data
        self.arf_specresp = np.array(arf['SPECRESP'])
        self.E_arf = np.array(arf['ENERG_LO'] + arf['ENERG_HI'])/2
        
        self.vign_on_arf_bins = np.zeros((self.theta.shape[0],
                                    E_arf.shape[0]))

        for i in range(self.theta.shape[0]):
            coeffs = quadratic_from_bin_averages(self.vign_E_bins, 
                                                 self.vignet_integre[i], 
                                                 boundary = 'natural')
            
            self.vign_on_arf_bins[i] = eval_quadratic_piecewise(self.vign_E_bins, 
                                                           coeffs, 
                                                           E_arf) 
        
        
    def load_event(self,
                   event_file_path,
                   grades_to_keep = [1]):
        
        self.hdu_evt = fits.open('/xifu/home/mola/XIFU_Sims_Turbulence_NewConfig/Observation5/Y+0_Z+0/Y+0_Z+0_125ks/events.fits')
        self.evt_data = self.hdu_evt[1].data
        self.evt_data = self.evt_data[np.isin(self.evt_data['GRADING'], grades_to_keep)] 

        
        # Angular distance from center in arcminutes
        self.evt_theta = np.degrees(
                                            np.arccos(
                                                np.cos(np.radians(self.evt_data['DEC'])) * np.cos(np.radians(self.evt_data['RA']))
                                                    ) 
                                            ) * 60
        
    def make_arfs_and_spectra(self, 
                              path,
                              numproc = 1):
        
        results = Parallel(n_jobs=numproc, 
                           backend="loky")(
                                delayed(self.make_single_arf_and_spectrum)(bin_number) for bin_number in tqdm_notebook(np.arange(self.binning.nb_bins), 
                                                                                                                                total = self.binning.nb_bins, 
                                                                                                                                desc = 'Making arfs and spectra'
                                                                                                                                              )
                                )
    
    def make_single_arf_and_spectrum(self, 
                                     bin_number, 
                                     arf_save_name = 'region_200/spectra/spec_',
                                     evt_save_name = 'region_200/spectra/spec_'):
        
        pixels_in_region=self.binning.binning_dict[bin_number][0]
        interpolated_vignet = np.zeros_like(E_arf)
        counts_tot = 0
        for i in range(0, len(pixels_in_region)):
            
            # Real pixel number
            pix_num=pixels_in_region[i] 
            
            
            #Compute theta
            theta_pix= np.mean(self.evt_theta[self.evt_data['PIXID'] == pix_num])

            # Weight with counts
            counts_in_band=np.sum(self.evt_data['PIXID'] == pix_num)
            counts_tot += counts_in_band

            # Broadcast interpolation on axis of energies
            func_to_vmap = lambda vignet_pt : jnp.interp(theta_pix, theta, vignet_pt) 
            vmapped_interp = jax.vmap(func_to_vmap)
            interpolated_vignet += vmapped_interp(vign_on_arf_bins.T) * counts_in_band 

        interpolated_vignet /= counts_tot
        print(interpolated_vignet)
        
        #Save ARF
        self.hdu_arf[1].data['SPECRESP'] = self.arf_specresp * np.array(interpolated_vignet)
        self.hdu_arf.writeto(arf_save_name + '{}.arf'.format(bin_number), overwrite=True)
        
        
        self.hdu_evt['EVENTS'].header['ANCRFILE']='/xifu/home/mola/xifu_cluster_sim/src/' + arf_save_name + '{}.arf'.format(bin_number)
        self.hdu_evt['EVENTS'].header['RESPFILE']= '/xifu/home/mola/XIFU_Sims_Turbulence_NewConfig/RMF.rmf'
        
        self.hdu_evt[1].data = self.evt_data[np.isin(self.evt_data['PIXID'],
                                                     pixels_in_region)]
        
        output_evt_file = evt_save_name + '{}.evt'.format(bin_number)
        self.hdu_evt.writeto(output_evt_file, overwrite=True)
        
        #Make spectrum using makespec
        pha_filename = output_evt_file.replace('.evt','.pha')
        arguments=["makespec",
                   "EvtFile=%s" % '/xifu/home/mola/xifu_cluster_sim/src/' + output_evt_file,
                   "Spectrum=%s" % '/xifu/home/mola/xifu_cluster_sim/src/' + pha_filename,
                   "clobber=T"]
        print("Calling:",arguments)
        subprocess.check_call(arguments)
        
```
