```python
def load_model(infile, pars_only=False):
    '''
    Retourne le model defini dans le fichier infile
    - infile (string) .xcm file
    - pars_only (boolean) option to return paramaters only
    '''
    #On recupere le nom du modele
    model_file = open(infile,'r')
    line = model_file.readline()
    while 'model' not in line:
        if 'cosmo' in line:
            line = line.replace('\n','')
            line = line.replace('cosmo ','')
            xs.Xset.cosmo = ' '.join(line.split(','))
            line = model_file.readline()
        if 'statistic' in line:
            line = model_file.readline()
            continue
        if 'method' in line:
            line = model_file.readline()
            continue
        if 'delta' in line:
            line = model_file.readline()
            continue
        if 'systematic' in line:
            line = model_file.readline()
            continue
        if 'abund' in line:
            line = line.replace('\n','')
            line = line.replace('abund ','')
            xs.Xset.abund = ' '.join(line.split(','))
            line = model_file.readline()
        if 'xsect' in line:
            line = line.replace('\n','')
            line = line.replace('xsect ','')
            xs.Xset.xsect = ' '.join(line.split(','))
            line = model_file.readline()
        
    line = line.replace('\n','')
    line = line.replace('model','')
    
    if not pars_only:
        m = xs.Model(line)
        nb_para = m.nParameters
        #On assigne les parametres
        i=1
        line = model_file.readline()
        while (line and i<=nb_para):
            line = line.replace('\n','')
            param_list = line.split()
            param_list = list(map(float,param_list))
            m(i).values =  param_list
            line=model_file.readline()
            i+=1
        
        return m
    
    else:
        mod_name=line
        Pars=[]
        Frozen=[]
        i=1
        line = model_file.readline()
        while (line):
            line = line.replace('\n','')
            param_list = line.split()
            param_list = list(map(float,param_list))
            Pars.append(param_list[0])
            if param_list[1]<0:
                Frozen.append(i)
            line=model_file.readline()
            i+=1
        
        return mod_name, Pars, Frozen
```

```python
def shakefit(dchi,model):
    nopar = model.nParameters
    xs.Fit.nIterations=100
    pattern = re.compile(r'F........')
    pattern2 = re.compile(r'.T.......')
    
    for j in range(nopar):
        if (model(j+1).frozen):
            continue 
        doerror=1
        delchi=dchi
        while (doerror == 1) :
            error_command = 'stopat 100 0.01 max 20.0 %.2f %d' % (delchi,j+1) #changed 20 to 50 (Edo)
            #print error_command
            xs.Fit.error(error_command)
            error_out = model(j+1).error[2]
            if (pattern.match(error_out)):
                doerror=0
            else:
                xs.Fit.perform()
            
            if (pattern2.match(error_out)):
                delchi*=2
```

```python
def generate_output_string(m,emin,emax,do_error=False,delchi=1.,ignore_flux=False):
    '''Generates the output string containing the fit results
    
    Arguments:
        - m: current model
        - emin,emax: energy range for flux computation
        - do_error: option to also retrieve the error values
        - delchi: delta chi2 value for error computation (currently only used for flux computation, will ask Etienne if this makes sense)
    
    Returns: output_string
    '''
    #On recupere le flux total, les erreurs et chi2 et on ecrit dans la sortie
    xs.AllModels.calcFlux("%.1f %.1f" % (emin,emax))
    s = xs.AllData(1)
    flux = s.flux[0]
    sortie = ',%g' % flux
    for k in range(1,m.nParameters+1):
        if not (m(k).frozen) and do_error:
            #Fit.error('%f %d' % (delchi,k))
            valeur = m(k).values[0]
            valeur_min = m(k).error[0]
            valeur_max = m(k).error[1]
        elif do_error:
            valeur = m(k).values[0]
            valeur_min = valeur
            valeur_max = valeur
        else:
            valeur = m(k).values[0]

        if do_error:
            sortie+=',%g,%g,%g' % (valeur,valeur_min,valeur_max)
        else:
            sortie+=',%g' % (valeur)

    #Calculate and save rchi2
    chi2 = xs.Fit.statistic
    dof = xs.Fit.dof
    rchi2 =chi2/dof
    sortie+=",%g" % rchi2

    if not (ignore_flux):
        d_val = dict_valeurs(m)
        expr = m.expression
        #Calcul du flux de chaque composante
        liste_comp = m.componentNames
        d_comp = dict_composantes(m)
        expr = m.expression
        for comp_name in liste_comp:
            if not (comp_name in ['wabs','fabs']):
                m = ajouter_cflux(d_val,expr,comp_name,d_comp,3.,30.)
                xs.Fit.perform()
                indice_flux = d_comp[comp_name][0] + 2
                Fit.error('%f %d' % (delchi,indice_flux))
                flux = m(indice_flux).values[0]
                flux_min = m(indice_flux).error[0]
                flux_max = m(indice_flux).error[1]
                sortie+=',%g,%g,%g' % (flux,flux_min,flux_max)
    return sortie
```

```python
def launch_fit_lite(filename,
                    model_filename,
                    outdir, 
                    arf_file=None, 
                    delchi = 1.,
                    silent=0,
                    ignore_flux=0,
                    skip_energy_range=True,
                    emin=0.2,
                    emax=12.,
                    systematics=0.,
                    logfile=None,
                    print_results=True,
                    ignore_bad=True,
                    do_error=True, 
                    background=False, 
                    pix_in_region=1, 
                    astro_bkg_model=None, 
                    cosmic_bkg_model=None, 
                    stat="cstat",
                    NXB_requirement=0.005):
    '''
    Fit un spectre et affiche les resultats du fit (lite version)
    Arguments:
        - filename: path to the spectrum file
        - model_filename: path to the model file
        - outdir: path to the outfile directory
        - arf_file: path to arf if vignetting
        - delchi: value of the delta chi to look for during error computation
        - silent: option to turn off console prints
        - ignore_flux: option to ignore flux computation
        - skip_energy_range: option to skip to energy range limitation (typically useful if the channels should be selected using bad flag from binning stage - see ignore_bad option)
        - emin,emax: energy range on which to fit and compute the total flux
        - logfile: path to optional log file
        - print_results: option to print on console the final results of the fit
        - ignore_bad: option to ignore channels flagged as bad during previous binning process
        - do_error: option to compute the errors on each free parameter
        - background: option to add a NXB option to the fit
        - pix_in_region: number of pixels in the region (useful to multiply a given norm)
        - astro_bkg_model, cosmic_bkg_model (string): path to astrophysical and cosmic x-ray background models
        - stat (string): to chose fitting method
    '''
    
    #Only do loading phase if this is the first time that this function has been called inside
    #a potential nested fit
    if (silent):    
        xs.Xset.chatter = 0
    else:
        xs.Xset.chatter = 10
    xs.AllChains.clear()
    xs.AllData.clear()
    xs.AllModels.clear()
    m = load_model(model_filename)
    
    #Fichiers log et ps
    if logfile=='default':
        logfile = outdir + '/' + filename.replace('.pha','.log')
    
    #On lance le logging
    if logfile!=None:
        xs.Xset.openLog(logfile)
            
    s=xs.Spectrum(filename)
    
    #Account for vignetting
    if arf_file is not None: 
        print("Replacing arf with %s" %arf_file)
        s.response.arf=arf_file
     
    if background: #Option to run the background
        ##NXB##
        print(arf_file, pix_in_region, 'pixels considered')
        NXB_level=NXB_requirement*(pitch)**2*pix_in_region #Value in cts/s/keV for a given pixel times the number of pixels /!\ HARDCODED PITCH and NXB level
        s.multiresponse[1]=s.multiresponse[0].rmf
        print(arf_file, 'NXB level :', NXB_level)
        m2=xs.Model("pow", modName="nxb", sourceNum=2, setPars={1:0,2:NXB_level})
        m2(1).frozen=True #constant
        #m2(2).frozen=True #constant
                
        ##AXB##
        if astro_bkg_model is not None:
            s.multiresponse[2]=s.multiresponse[0].rmf
            s.multiresponse[2].arf=s.multiresponse[0].arf
            axb_model, axb_pars, axb_frozen=load_model(astro_bkg_model, pars_only=True)
            m3=xs.Model(axb_model, modName="axb", sourceNum=3, setPars=axb_pars)
            for j in axb_frozen: #Freeze all parameters (assumed perfectly known)
                m3(j).frozen=True 
            #Set the 'right norm
            for j in range(m3.nParameters):    
                if m3(j+1).name=='norm':
                    m3.setPars({j+1:m3(j+1).values[0]*pix_in_region*aminconv**2})
                
        ##CXB##
        if cosmic_bkg_model is not None:
            s.multiresponse[3]=s.multiresponse[0].rmf
            s.multiresponse[3].arf=s.multiresponse[0].arf
            cxb_model, cxb_pars, cxb_frozen=load_model(cosmic_bkg_model, pars_only=True)
            m4=xs.Model(cxb_model, modName="cxb", sourceNum=4, setPars=cxb_pars)
            for j in cxb_frozen: #Freeze all right parameters
                m4(j).frozen=True
            m4.setPars({m4.nParameters:m4(m4.nParameters).values[0]*pix_in_region*aminconv**2}) #Set 'right' norm
        
    #Fit a proprement parler
    if systematics:
        xs.AllModels.systematic = 0.005
    if not skip_energy_range:
        xs.AllData.ignore('0.0-%.1f' % emin)
        xs.AllData.ignore('%.1f-**' % emax)
    if ignore_bad:
        xs.AllData.ignore('bad')
    xs.Fit.nIterations = 1000   
    xs.Fit.statMethod = stat
    xs.Fit.query = 'yes'

    xs.Fit.perform()
    xs.Fit.perform()
    xs.Fit.perform()
        
    #Call shakefit to avoid local minima in case error computation has been requested
    if do_error:
        shakefit(delchi,m)
    
    #Return chi2 if this is a nested fit
    chi2 = xs.Fit.statistic

    #Get fit results
    sortie = generate_output_string(m, emin, emax, do_error, delchi, ignore_flux)
   
    #Fin du logging
    if logfile!=None:
        xs.Xset.closeLog()    
        
    if print_results:
        sys.stdout.write(sortie)
    
    return sortie
```

```python
from joblib import Parallel, delayed
import joblib

from tqdm.notebook import tqdm as tqdm_notebook
```

```python
class FitSpectra():
    def __init__(self,
                model_file,
                binning,
                background=False,
                E_min = 0.2,
                E_max = 12.,
                NXB_requirement = 5e-3,
                astro_bkg_model = None, 
                cosmic_bkg_model = None,
                xifu_pixel_size_m = 317e-6,
                athena_focal_length = 12):
        
        self.model_file = model_file
        self.binning = binning
        self.E_min = E_min
        self.E_max = E_max
        self.background = background
        self.NXB_requirement = NXB_requirement
        self.astro_bkg_model = astro_bkg_model
        self.cosmic_bkg_model = cosmic_bkg_model
        self.xifu_pixel_surface_m = (xifu_pixel_size_m*100)**2
        self.xifu_pixel_size_arcmin = (xifu_pixel_size_m/athena_focal_length*units.radian).to(units.arcmin).value
        
    def fit_spectrum(self,
                    pha_file,
                    arf_file,
                    pix_in_region,
                    log_fitting = True,
                    chatter = 10):
              
        print('current_pha_file :',pha_file)
        xs.Xset.chatter = chatter
        xs.AllChains.clear()
        xs.AllData.clear()
        xs.AllModels.clear()

        #xs.Xset.restore(self.model_file)
        #m = xs.AllModels(1)
        m = load_model(self.model_file)

        logfile = pha_file.replace('.pha','.log')
        if log_fitting :
            xs.Xset.logChatter = 10
            xs.Xset.openLog(logfile)

        s = xs.Spectrum(pha_file)
        s.response.arf = arf_file

        # Option to run the background
        if self.background: 

            ## NXB 

            #Value in cts/s/keV for a given pixel times the number of pixels
            NXB_level = self.NXB_requirement * self.xifu_pixel_surface_m * pix_in_region
            s.multiresponse[1]=s.multiresponse[0].rmf
            m2=xs.Model("pow", modName="nxb", sourceNum=2, setPars={1:0,2:NXB_level})
            m2(1).frozen=True #constant

            ## AXB
            if self.astro_bkg_model is not None:
                s.multiresponse[2]=s.multiresponse[0].rmf
                s.multiresponse[2].arf=s.multiresponse[0].arf

                axb_model, axb_pars, axb_frozen = load_model(self.astro_bkg_model, pars_only=True)
                m3=xs.Model(axb_model, modName="axb", sourceNum=3, setPars=axb_pars)

                # Freeze parameters
                for j in axb_frozen: 
                    m3(j).frozen=True 

                # Multiply the norm by the angular size of the region
                region_size_arcmin = self.xifu_pixel_size_arcmin**2 * pix_in_region
                m3.setPars({4:m3(4).values[0]*region_size_arcmin})
                m3.setPars({9:m3(9).values[0]*region_size_arcmin})

            ## CXB
            if self.cosmic_bkg_model is not None:
                s.multiresponse[3]=s.multiresponse[0].rmf
                s.multiresponse[3].arf=s.multiresponse[0].arf

                cxb_model, cxb_pars, cxb_frozen=load_model(self.cosmic_bkg_model, pars_only=True)
                m4=xs.Model(cxb_model, modName="cxb", sourceNum=4, setPars=cxb_pars)

                # Freeze parameters
                for j in cxb_frozen: 
                    m4(j).frozen=True

                # Multiply the norm by the angular size of the region
                region_size_arcmin = self.xifu_pixel_size_arcmin**2 * pix_in_region
                m4.setPars({3:m4(3).values[0]*region_size_arcmin})

        # Fitting
        xs.AllData.ignore('0.0-%.1f' % self.E_min)
        xs.AllData.ignore('%.1f-**' % self.E_max)

        xs.AllData.ignore('bad')
        xs.Fit.statMethod = "cstat"

        xs.Fit.nIterations = 1000   
        xs.Fit.query = 'yes'

        xs.Fit.perform()
        xs.Fit.perform()
        xs.Fit.perform()

        # Call shakefit to avoid local minima in case error computation has been requested
        shakefit(1,m)

        # Return chi2 if this is a nested fit
        chi2 = xs.Fit.statistic

        # Get fit results
        m.show()

        pars_dict = {}
        for k in range(1,1+m.nParameters):
            pars_dict[m(k).name] = [m(k).values[0], m(k).error[0], m(k).error[1]]

        pars_dict['chi2'] = chi2
        
        # End logging
        if log_fitting :
            xs.Xset.closeLog()  
        return pars_dict
    
    def fit_all_spectra(self,
                        spectra_path,
                        numproc=1,
                        log_fitting=False,
                        chatter = 0):
        
        # Test value of 3
        n_bins = 3 #self.binning.nb_bins
        args_list = zip([spectra_path + 'spec_{}_binned.pha'.format(k) for k in range(1,n_bins+1)],
                        [spectra_path + 'spec_{}.arf'.format(k) for k in range(1,n_bins+1)],
                        [len(self.binning.binning_dict[k][0]) for k in range(1,n_bins+1)])
        
        results = Parallel(n_jobs=numproc, 
                           backend="loky")(
                                delayed(self.fit_spectrum)(pha_file, 
                                                           arf_file,
                                                           nb_pix_in_bin,
                                                           log_fitting = log_fitting,
                                                           chatter = chatter) for pha_file, arf_file, nb_pix_in_bin in tqdm_notebook(args_list, 
                                                                                                                               total = n_bins, 
                                                                                                                               desc = 'Fitting spectra'
                                                                                                                               )
                                )
        
        return results
        
        
```

```python
solver.plot_ppc("prior", component_names=["Total"], figsize=(4.5, 4.5));
```

```python
from embedding.trainable import AutoencoderEmbedding
from embedding.summary import GlobalSummaryEmbedding, LocalSumWithExtra
from embedding import MultipleEmbedding

summary_embedding = MultipleEmbedding([
    GlobalSummaryEmbedding(),
    LocalSumWithExtra(np.linspace(low_energy, high_energy, 31)),
])

aa_embedding = AutoencoderEmbedding(latent_dim=32, retrain_from_scratch=False)
```

```python
posterior = solver.run(
    num_simulations_per_round=[20_000, 5_000, 5_000], # 3 rounds with 20_000, 5_000 and 5_000 simulations
    embedding=[summary_embedding, aa_embedding, aa_embedding] # First round will use summary statistics, second and third rounds use an autoencoder
)
```

```python
# Plot the posterior predictive distribution using the exact sampler
component_names = ["Total", "blackbodyrad", "powerlaw"]
fig = solver.plot_ppc("exact_sampler", component_names=component_names, y_lim=(1e-6, 1e-2), figsize=(4.5,4.5), n_samples=100);
```
