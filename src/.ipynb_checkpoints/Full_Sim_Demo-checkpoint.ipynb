### Plotting func

```python
def plotmaps(input_map1,
            input_map2_name = 'input_norm',
            cmap = cmr.ember,
            title = 'NORM',
            minval = None,
            maxval = None,
             plotlog = False,
             plotdiff_not_ratio = False):
    
    path = '/xifu/home/mola/XIFU_Sims_Turbulence_NewConfig/Observation5/Y+0_Z+0/Y+0_Z+0_125ks/region_200/'

    input_map2 = np.array(fits.getdata(path + 'input_maps/'+input_map2_name+'.fits'), dtype = 'float64')
    
    if plotlog : 
        input_map1 = np.log10(input_map1)
        input_map2 = np.log10(input_map2)


    if minval == None :
        minval = min(input_map1.min(), input_map2.min())
    if maxval == None : 
        maxval = max(input_map1.max(), input_map2.max())
        
    fig, ax = plt.subplots(1,3,figsize = (15,6), sharey = True)
          
    im = ax[1].imshow(input_map1, cmap = cmap, vmin = minval, vmax = maxval)
    plt.colorbar(im,fraction=0.046, pad=0.04)
    ax[1].set_title('INPUT ME')
    ax[1].set_xticks(())
    ax[1].set_yticks(())

    im = ax[0].imshow(input_map2, cmap = cmap, vmin = minval, vmax = maxval)
    plt.colorbar(im,fraction=0.046, pad=0.04) 
    ax[0].set_title('INPUT OLD')
    ax[0].set_ylabel(title)
    ax[0].set_xticks(())
    ax[0].set_yticks(())
    
    if not plotdiff_not_ratio :
        im = ax[2].imshow(input_map1/input_map2, cmap = cmr.iceburn, vmin = 0.99, vmax = 1.01)
        plt.colorbar(im,fraction=0.046, pad=0.04) 
        ax[2].set_title('RATIO')
        ax[2].set_xticks(())
        ax[2].set_yticks(())
    
    else : 
        diff = input_map1 - input_map2
        im = ax[2].imshow(diff, cmap = cmr.iceburn, vmin = -50, vmax = 50)
        plt.colorbar(im,fraction=0.046, pad=0.04) 
        ax[2].set_title('DIFFERENCE')
        
    plt.subplots_adjust()
    
    #return input_map1, input_map2
```

```python
def plot_output_maps(input_map1,
            input_map2_name = 'output_norm',
            cmap = cmr.ember,
            title = 'NORM',
            minval = None,
            maxval = None,
             plotlog = False,
             plotdiff_not_ratio = False):
    
    path = '/xifu/home/mola/XIFU_Sims_Turbulence_NewConfig/Observation5/Y+0_Z+0/Y+0_Z+0_125ks/region_200/'

    input_map2 = np.array(fits.getdata(path + 'output_maps/'+input_map2_name+'.fits'), dtype = 'float64')
    
    if plotlog : 
        input_map1 = np.log10(input_map1)
        input_map2 = np.log10(input_map2)


    if minval == None :
        minval = min(input_map1.min(), input_map2.min())
    if maxval == None : 
        maxval = max(input_map1.max(), input_map2.max())
        
    fig, ax = plt.subplots(1,3,figsize = (15,6), sharey = True)
          
    im = ax[1].imshow(input_map1, cmap = cmap, vmin = minval, vmax = maxval)
    plt.colorbar(im,fraction=0.046, pad=0.04)
    ax[1].set_title('OUTPUT ME')
    ax[1].set_xticks(())
    ax[1].set_yticks(())

    im = ax[0].imshow(input_map2, cmap = cmap, vmin = minval, vmax = maxval)
    plt.colorbar(im,fraction=0.046, pad=0.04) 
    ax[0].set_title('OUTPUT OLD')
    ax[0].set_ylabel(title)
    ax[0].set_xticks(())
    ax[0].set_yticks(())
    
    if not plotdiff_not_ratio :
        im = ax[2].imshow(input_map1/input_map2, cmap = cmr.iceburn, vmin = 0.75, vmax = 1.25)
        plt.colorbar(im,fraction=0.046, pad=0.04) 
        ax[2].set_title('RATIO')
        ax[2].set_xticks(())
        ax[2].set_yticks(())
    
    else : 
        diff = input_map1 - input_map2
        im = ax[2].imshow(diff, cmap = cmr.iceburn, vmin = -50, vmax = 50)
        plt.colorbar(im,fraction=0.046, pad=0.04) 
        ax[2].set_title('DIFFERENCE')
        
    plt.subplots_adjust()
    
    #return input_map1, input_map2
```

```python
plot_output_maps(best_fit_norm,minval = 0., plotlog = False)

plot_output_maps(best_fit_kT,
            input_map2_name = 'output_kt',
            cmap = cmr.ember,
            title = 'TEMPERATURE',
            minval = 5.,
            maxval = 6.,
             plotlog = False)

plot_output_maps(best_fit_Z,
            input_map2_name = 'output_abundanc',
            cmap = cmr.ember,
            title = 'ABUNDANCE',
            minval = 0.1,
            maxval = 0.9,
             plotlog = False)

plot_output_maps(best_fit_z,
            input_map2_name = 'output_redshift',
            cmap = cmr.iceburn,
            title = 'REDSHIFT',
            minval = 0.099,
            maxval = 0.101,
             plotlog = False)

plot_output_maps(best_fit_broad,
            input_map2_name = 'output_velocity',
            cmap = cmr.ember,
            title = 'REDSHIFT',
            minval = 50,
            maxval = 300,
             plotlog = False)
```
